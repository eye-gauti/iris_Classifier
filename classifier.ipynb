#### Data visualisiation

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0, 0].hist(df['SepalLengthCm'], bins=20, color='skyblue')
axs[0, 0].set_title('Sepal Length')

axs[0, 1].hist(df['SepalWidthCm'], bins=20, color='lightgreen')
axs[0, 1].set_title('Sepal Width')

axs[1, 0].hist(df['PetalLengthCm'], bins=20, color='salmon')
axs[1, 0].set_title('Petal Length')

axs[1, 1].hist(df['PetalWidthCm'], bins=20, color='orange')
axs[1, 1].set_title('Petal Width')


```

```python
colors = ['skyblue', 'lightgreen', 'salmon']
species = ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Iris Feature Scatter Plots by Species', fontsize=16)

# Sepal Length vs Sepal Width
for i in range(3):
    x = df[df['Species'] == species[i]]
    axs[0, 0].scatter(x['SepalLengthCm'], x['SepalWidthCm'], c=colors[i], label=species[i])
axs[0, 0].set_xlabel('Sepal Length')
axs[0, 0].set_ylabel('Sepal Width')
axs[0, 0].legend()
axs[0, 0].set_title('Sepal Length vs Sepal Width')

# Petal Length vs Petal Width
for i in range(3):
    x = df[df['Species'] == species[i]]
    axs[0, 1].scatter(x['PetalLengthCm'], x['PetalWidthCm'], c=colors[i], label=species[i])
axs[0, 1].set_xlabel('Petal Length')
axs[0, 1].set_ylabel('Petal Width')
axs[0, 1].legend()
axs[0, 1].set_title('Petal Length vs Petal Width')

# Sepal Width vs Petal Width
for i in range(3):
    x = df[df['Species'] == species[i]]
    axs[1, 0].scatter(x['SepalWidthCm'], x['PetalWidthCm'], c=colors[i], label=species[i])
axs[1, 0].set_xlabel('Sepal Width')
axs[1, 0].set_ylabel('Petal Width')
axs[1, 0].legend()
axs[1, 0].set_title('Sepal Width vs Petal Width')

# Sepal Length vs Petal Length
for i in range(3):
    x = df[df['Species'] == species[i]]
    axs[1, 1].scatter(x['SepalLengthCm'], x['PetalLengthCm'], c=colors[i], label=species[i])
axs[1, 1].set_xlabel('Sepal Length')
axs[1, 1].set_ylabel('Petal Length')
axs[1, 1].legend()
axs[1, 1].set_title('Sepal Length vs Petal Length')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

As we can see from the various scatter plots and histograms, the Iris-setosa class is clearly distinguishable from the other two classes. In particular, the petal length and petal width features show a clear separation between all three classes, with little to no overlap. However, in the plots involving sepal length and sepal width, there is significant overlap between Iris-versicolor and Iris-virginica, which makes it more challenging to differentiate between them based on those features alone
